```python
import requests
import pandas as pd
```

```python
API_KEY = 'AIzaSyC-HW6lb8biPigaulcWJVi70NnHuog0dYU'
VIDEO_ID = 'WOHFlwYV210'
BASE_URL = 'https://www.googleapis.com/youtube/v3/videos'
params = {
    'part': 'snippet,contentDetails,statistics',
    'id': VIDEO_ID,
    'key': API_KEY,
    'maxResults': 1000 
}
```

```python
def fetch_youtube_data():
    all_data = []
    next_page_token = None

    while True:
        if next_page_token:
            params['pageToken'] = next_page_token

        response = requests.get(BASE_URL, params=params)
        
        if response.status_code != 200:
            print(f"Error: Received status code {response.status_code}")
            print(response.json())
            break

        result = response.json()
        print(f"Response: {result}")  # Debug: Print the full response

        items = result.get('items', [])
        if not items:
            print("No items found in the response.")
            break

        all_data.extend(items)

        next_page_token = result.get('nextPageToken')
        if not next_page_token:
            break

    return all_data
```

```python
API_KEY = 'AIzaSyC-HW6lb8biPigaulcWJVi70NnHuog0dYU'
VIDEO_ID = 'WOHFlwYV210'
BASE_URL = 'https://www.googleapis.com/youtube/v3/commentThreads'
params = {
    'part': 'snippet',
    'videoId': VIDEO_ID,
    'key': API_KEY,
    'maxResults': 1000  
}
```

```python
def fetch_comments():
    comments_data = []
    next_page_token = None

    while True:
        if next_page_token:
            params['pageToken'] = next_page_token

        response = requests.get(BASE_URL, params=params)
        
        if response.status_code != 200:
            print(f"Error: Received status code {response.status_code}")
            print(response.json())
            break

        result = response.json()

        for item in result.get('items', []):
            snippet = item.get('snippet', {})
            top_level_comment = snippet.get('topLevelComment', {})
            comment_snippet = top_level_comment.get('snippet', {})

            comment_data = {
                'user_name': comment_snippet.get('authorDisplayName'),
                'user_image': comment_snippet.get('authorProfileImageUrl'),
                'user_location': comment_snippet.get('authorChannelUrl'),  # Location is not typically provided
                'comment': comment_snippet.get('textDisplay')
            }
            comments_data.append(comment_data)

        next_page_token = result.get('nextPageToken')
        if not next_page_token:
            break

    return comments_data

comments = fetch_comments()
df = pd.DataFrame(comments)
df
```
